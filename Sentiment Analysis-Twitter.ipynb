```python
raw_tweets = []
for tweets in j_ans:
    raw_tweets.append(tweets['text'])
raw_tweets

raw_string=''.join(raw_tweets)
no_links = re.sub(r'http\S+', '', raw_string)
no_unicode = re.sub(r"\\[a-z][a-z]?[0-9]+", '', no_links)

no_special_characters = re.sub('[^A-Za-z ]+', '', no_unicode)
raw_tweets
```

```python
sentiment_dict={'neg': 0, 'neu': 0, 'pos': 0, 'compound': 0}

for tweet in raw_tweets:
    print(tweet)
    
    senti=SentimentIntensityAnalyzer()
    sent_dict=senti.polarity_scores(tweet)
    print(sent_dict)
    
    x=sorted(sent_dict, key=(lambda key:sent_dict[key]), reverse=True)
    print("The overall sentiment is:",x[0])
    
    sentiment_dict[x[0]] +=1
    print(sentiment_dict)
    print("\n")
```

```python
import json
import requests
import re
from requests_oauthlib import OAuth1
from textblob import TextBlob         #for checking the sentiment of each tweet
from collections import Counter       #for counting each sentiment


api_key = 'Z0P4OidK4cKttEmzDdQRSckpa' # enter api key
api_secret_key = 'rqpAf5m3pbnoanYR1Z5BLYFkksiuxhwOQzhv8PeVmHsa7rXySR'  #enter api secret key
auth = OAuth1(api_key, api_secret_key)
screen_name = 'realDonaldTrump' #search keyword  #realDonaldTrump
url="https://api.twitter.com/1.1/statuses/user_timeline.json?screen_name="+screen_name+"&count=200"
ans = requests.get(url, auth=auth)
jtweet = json.loads(ans.text)

twi_li=[]
for x in range(1,200):
	#print (x)
	a=jtweet[x]['text']
	clean_var= ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) |(\w+:\/\/\S+)", " ", a).split())
	a1=TextBlob(clean_var)
	print(a1.sentiment.polarity)
	if (a1.sentiment.polarity > 0):
		twi_li.append("positive")
	elif (a1.sentiment.polarity < 0):
		twi_li.append("negative")
	else:
		twi_li.append("neutral")
twi_li_count=Counter(twi_li)
#print(twi_li_count)
b=[]
for x in twi_li_count: 
	b.append(twi_li_count[x])
	#print(b)
key_max=max(b)
max_words = [key for key,value in twi_li_count.items() if value==key_max]

#print(max_words)

print("The overall sentiment analysis: "+ str(*max_words)) # "*" is used to remove brackets

```
